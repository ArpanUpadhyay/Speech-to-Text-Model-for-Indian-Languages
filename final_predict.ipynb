```python
MODEL_ID="jonatasgrosman/wav2vec2-large-xlsr-53-english"
from huggingsound import SpeechRecognitionModel
from transformers import Wav2Vec2Processor,Wav2Vec2ForCTC
model = Wav2Vec2ForCTC.from_pretrained(MODEL_ID)
processor = Wav2Vec2Processor.from_pretrained(MODEL_ID)
```

```python
#data=speech_array.flatten()
```

```python
#data.shape
```

```python
num_channels, _ = speech_array.shape
if num_channels == 1:
    speech_array=speech_array[0]
else:
    raise ValueError("Waveform with more than 1 channels are not supported.")
```

```python
import torch
```

```python
inputs = processor(speech_array, sampling_rate=16_000, return_tensors="pt", padding=True)
p=model(inputs.input_values, attention_mask=inputs.attention_mask)
logits=p.logits
p_ids=torch.argmax(logits, dim=-1)[0]
predicted_sentence=processor.decode(p_ids)
predicted_sentence
```
